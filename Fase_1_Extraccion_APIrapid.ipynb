```python
lista_generos =["Drama", "Comedy", "Action", "Fantasy", "Horror", "Mystery", "Romance", "Thriller"]
#Iremos sacando un archivo por cada genero 
def sacar_datos_api(): #url o endpoint
    """
    Realiza una llamada a una API utilizando la URL proporcionada, para sacar los datos de las películas

    Parameters:
    -----------
    url (str): La URL de la API que se va a llamar. url = f"https://moviesdatabase.p.rapidapi.com/titles?year={año}&genre=Comedy&page={pagina}&titleType={tipo}&limit=50"

    Returns:
    --------
    dict or None: Un diccionario con los datos de respuesta de la API si la llamada fue exitosa (código de estado 200).
                None si la llamada falló o no se pudo autenticar correctamente.
    """

    lista_pelis = []
    #Genero en mayúscula la primera! 
    #Cambiar a "movies" para sacar las peliculas
    for tipo in ["short"]:
        #Iremos cambiando los años para mover el rango (hemos hecho tandas de 10 en 10)
        for año in range(1990,1991): 
            #El máximo de películas que nos devuelve es de 50 con 50 películas en cada página.
            for pagina in range(1, 2): 
                headers = {
                "X-RapidAPI-Key": "08e908296bmsh570bcd982cd4403p180575jsn94fec8140424",
                "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
                    }    
                url = f"https://moviesdatabase.p.rapidapi.com/titles?year={año}&genre=Comedy&page={pagina}&titleType={tipo}&limit=50"
                response = requests.get(url,headers=headers)
                json_response = response.json()
                try:
                    json_response["results"]
                except:
                    break
                lista_pelis.append(json_response)
                
    return lista_pelis
```

```python
#Sacamos diferentes archivos con la información que luego unimos en fases posteriores

def creando_lista_tuplas(list_pelis):

    lista_movie_tanda_comedy= []
    for peli in list_pelis:
        for i in peli["results"]:
            lista_todas_pelis = []
            tipo_peli = i['titleType']["text"]
            nombre_peli = i['titleText']["text"]
            id_peli =i['id']
            lista_todas_pelis.append(tipo_peli)
            lista_todas_pelis.append(nombre_peli)
            lista_todas_pelis.append(id_peli)
            
            try: 
                mes_estreno_peli = i["releaseDate"]["month"]
            except:
                mes_estreno_peli = ''
            lista_todas_pelis.append(mes_estreno_peli)
            try:
                ano_estreno_peli = i["releaseDate"]["year"]
            except:
                ano_estreno_peli = ''
            lista_todas_pelis.append(ano_estreno_peli)
            lista_movie_tanda_comedy.append(tuple(lista_todas_pelis))
        
    return lista_movie_tanda_comedy
```

```python
df_lista_movie_primera_tanda_comedy = pd.DataFrame(creando_lista_tuplas(lista_peliculas),columns = ["Tipo", "Título", "ID", "Mes Estreno",  "Año Estreno"])
df_lista_movie_primera_tanda_comedy
df_lista_movie_primera_tanda_comedy.to_csv("comedy_shorts_1990-2010.csv")
```
